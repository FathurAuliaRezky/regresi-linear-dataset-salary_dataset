```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# Membaca dataset
data = pd.read_csv('salary_dataset.csv')
```

Import dataset: Menggunakan Pandas, membaca file CSV yang berisi data salary_dataset.csv. Data ini terdiri dari dua kolom: no (nomor) dan salary (gaji).

```python
# Memisahkan fitur dan target
X = data['YearsExperience'].values.reshape(-1,1)
y = data['Salary'].values

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Pembuatan model: memisahkan fitur (YearsExperience) dan target (Salary) dari dataset, lalu membagi dataset menjadi data latih dan data uji. Kemudian, membuat model regresi linear menggunakan library sklearn. Menggunakan metode train_test_split untuk memisahkan data menjadi data latih dan data uji dengan rasio 80:20.

```python
# Membuat model regresi linear
model = LinearRegression()

# Melatih model
model.fit(X_train, y_train)

# Melakukan prediksi
y_pred = model.predict(X_test)
```

Membuat model regresi linear, latih model menggunakan data latih, dan lakukan prediksi menggunakan data uji

```python
# Menghitung akurasi
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = metrics.r2_score(y_test, y_pred)

print('Mean Squared Error:', mse)
print('R-squared:', r2)
```

Akurasi: menghitung akurasi model menggunakan metrik yang sesuai, misalnya mean squared error (MSE) dan coefficient of determination (R-squared). MSE menunjukkan seberapa jauh model dari data yang sebenarnya, sedangkan R-squared menunjukkan seberapa besar varians dari data yang dapat dijelaskan oleh model.

```python
# Melakukan prediksi untuk YearsExperience = 1.1, 11, dan 7
new_data = np.array([1.1, 11, 7]).reshape(-1, 1)
predictions = model.predict(new_data)
print('Predictions:', predictions)
```

Pengujian: melakukan prediksi untuk YearsExperience = 1.1, 11, dan 7 menggunakan model yang telah dibuat.

```python
# Membuat plot garis untuk data latih
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, model.predict(X_train), color='red')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Visualisasi: membuat plot garis untuk menunjukkan hubungan antara YearsExperience dan Salary pada data latih. Garis merah menunjukkan model regresi linear yang telah dibuat.
